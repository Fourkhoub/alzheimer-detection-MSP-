```python
# Lecture du csv

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/raw/alzheimers_disease_data.csv')
print(df.shape)
print(df.columns.tolist())
df.head()
```

```python
# Informations générales sur le dataset

print(" SHAPE ")
print(df.shape)

print("\n TYPES DES COLONNES ")
print(df.dtypes)

print("\n VALEURS MANQUANTES ")
print(df.isnull().sum().sort_values(ascending=False))

print("\n STATISTIQUES DESCRIPTIVES ")
df.describe()
```

```python
# Distribution de la variable cible
print(" VARIABLE CIBLE : Diagnosis ")
print(df['Diagnosis'].value_counts())
print(df['Diagnosis'].value_counts(normalize=True).round(3))
```

```python
# Visualisation de la distribution de la cible
plt.figure(figsize=(6, 4))
df['Diagnosis'].value_counts().plot(
    kind='bar',
    color=['#1D9E75', '#A32D2D'],
    edgecolor='white'
)
plt.title('Distribution Alzheimer (1) / Non-Alzheimer (0)')
plt.xticks([0, 1], ['Non-Alzheimer', 'Alzheimer'], rotation=0)
plt.ylabel('Nombre de patients')
plt.tight_layout()
plt.savefig('../reports/target_distribution.png', dpi=150)
plt.show()
```

```python
# Supprimer les colonnes inutiles

df = df.drop(columns=['PatientID', 'DoctorInCharge'])
print(f'Shape après suppression : {df.shape}')
```

```python
# Corrélations avec la variable cible

corr = df.corr()['Diagnosis'].drop('Diagnosis').sort_values(key=abs, ascending=False)

plt.figure(figsize=(10, 8))
corr.head(15).plot(kind='barh', color=['#A32D2D' if x > 0 else '#1D9E75' for x in corr.head(15)])
plt.title('Top 15 features corrélées avec Alzheimer')
plt.axvline(0, color='white', linewidth=0.8)
plt.tight_layout()
plt.savefig('../reports/feature_correlations.png', dpi=150)
plt.show()

print("\nTop 10 corrélations :")
print(corr.head(10))
```

```python
# Distribution de l'âge par diagnostic
plt.figure(figsize=(8, 5))
for diag, color, label in [(0, '#1D9E75', 'Non-Alzheimer'), (1, '#A32D2D', 'Alzheimer')]:
    df[df['Diagnosis'] == diag]['Age'].hist(
        alpha=0.6, bins=20, color=color, label=label
    )
plt.title("Distribution de l'âge par diagnostic")
plt.xlabel('Age')
plt.ylabel('Nombre de patients')
plt.legend()
plt.tight_layout()
plt.savefig('../reports/age_distribution.png', dpi=150)
plt.show()
```

```python
# Résumé EDA
print(" RÉSUMÉ EDA ")
print(f"Lignes : {df.shape[0]}")
print(f"Colonnes : {df.shape[1]}")
print(f"Valeurs manquantes : {df.isnull().sum().sum()}")
print(f"Non-Alzheimer (0) : {(df['Diagnosis']==0).sum()} ({(df['Diagnosis']==0).mean():.1%})")
print(f"Alzheimer (1) : {(df['Diagnosis']==1).sum()} ({(df['Diagnosis']==1).mean():.1%})")
print(f"\nTop 5 features corrélées :")
print(df.select_dtypes(include='number').corr()['Diagnosis'].drop('Diagnosis').abs().sort_values(ascending=False).head())
```
